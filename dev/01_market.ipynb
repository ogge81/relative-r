```python
import sys
sys.path.insert(0, "..")

from src.yfin.sector import Sector, SECTORS
from src.yfin.industry import Industry
from src.yfin.ticker import Ticker

from src.utils.files import add_to_json

for name in SECTORS:
    sector = Sector(name)
    indus = sector.industries

    if sector.overview is not None:
        print(sector.overview)

        # add_to_json(name, sector.overview, "sectors")

    if indus is not None:
        ikeys = indus.index.tolist()

        for ikey in ikeys:
            indust = Industry(ikey)

            indObj = indust.overview
            indObj["sector"] = name

            add_to_json(ikey, indObj, "industries")

            companies = indust.top_companies

            if companies is not None:
                symbols = companies.index.tolist()
                for symbol in symbols:

                    if symbol is not None:
                        ticker = Ticker(symbol)
                        add_to_json(symbol, {
                            "sector": name,
                            "industry": ikey,
                            "market_cap": ticker.fast_info["marketCap"],
                            "previous_close": ticker.fast_info["previousClose"],
                            "last_volume": ticker.fast_info["lastVolume"],
                        }, "assets")



# sect = Sector('utilities')
# indus = sect.industries
# ikeys = indus.index.tolist()

# print(ikeys[0])

# indust = Industry(ikeys[0])
# indust.top_companies
# sect.top_companies
```
